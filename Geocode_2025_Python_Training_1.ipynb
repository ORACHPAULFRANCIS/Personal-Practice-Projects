```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Read imported shapefile
gdf = gpd.read_file(r'/content/Data/District_2022.shp')
```

```python
# Replace DName2019 with Districts
gdf.rename(columns={'DName2019': 'Districts'}, inplace=True)
```

```python
# Plot geodataframe
gdf.plot()
```

```python
# Specify column to plot
fig, ax = plt.subplots(1, figsize=(10,10))
# Set title to The 15 Regions of Uganda
ax.set_title('The 15 Regions of Uganda')
# Plot the data
gdf.plot(ax=ax, column='F15Regions', legend=True, legend_kwds={'loc': 'upper left'})
leg = ax.get_legend()
leg.set_bbox_to_anchor((1.05, 0.5))
# Download our plot
plt.savefig('Uganda.png', dpi=300, bbox_inches='tight')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Set plot style
sns.set(style='whitegrid')

# Create the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='time', y='tip', data=tips, palette='Set2')

# Title and labels
plt.title('Tip Amount by Time of Day', fontsize=14)
plt.xlabel('Meal Time')
plt.ylabel('Tip Amount ($)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='size', y='tip', hue='time', data=tips, palette='coolwarm', s=100)

plt.title('Tip vs Party Size by Meal Time', fontsize=14)
plt.xlabel('Party Size')
plt.ylabel('Tip Amount ($)')
plt.legend(title='Meal Time')
plt.show()
```

```python
sns.lmplot(x='size', y='tip', hue='time', data=tips, palette='muted', aspect=1.5)
plt.title('Trend: Tip Amount vs Party Size by Time of Day')
```
