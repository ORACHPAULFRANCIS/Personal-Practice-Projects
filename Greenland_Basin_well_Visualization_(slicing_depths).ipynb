```python
# Importing required libraries for my exercise
import lasio
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the las file to Colab
from google.colab import files
las = files.upload()
```

```python
# Read the las file
las = lasio.read('greenland_GR.las')
las.sections.keys()
```

```python
las.sections['Version']
```

```python
for item in las.sections['Well']:
  print(f'Curve: ({item.mnemonic}) {item.descr}: \t {item.value}')
```

```python
for curve in las.curves:
  print(curve.mnemonic)
```

```python
well = las.df()
```

```python
well.head()
```

```python
well.info()
```

```python
well.describe()
```

```python
well.plot(figsize = (12, 6), title = 'Greenland Basin Gamma Ray (API)', ylabel = 'GR (API)')

# Since the depth is too large, the rendered curve is hard to interpret so
# I am going to slice different sections of the depth curve to get
# proper GR interpretation
```

```python
# Slice from depth 0 to 2000
well.loc[0:500, 'GR'].plot(
    figsize=(10, 4),
    title='Greenland Basin Gamma Ray (API)',
    ylabel='GR (API)'
)
```
