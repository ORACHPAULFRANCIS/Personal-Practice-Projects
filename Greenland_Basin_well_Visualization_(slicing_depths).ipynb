```python
# Importing required libraries for my exercise
import lasio
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the las file
las = lasio.read('greenland_GR.las')
las.sections.keys()
```

```python
las.sections['Version']
```

```python
for item in las.sections['Well']:
  print(f'Curve: ({item.mnemonic}) {item.descr}: \t {item.value}')
```

```python
for curve in las.curves:
  print(curve.mnemonic)
```

```python
well = las.df()
```

```python
well.head()
```

```python
well.info()
```

```python
well.describe()
```

```python
well.plot(figsize = (12, 6), title = 'Greenland Basin Gamma Ray (API)', ylabel = 'GR (API)')

# Since the depth is too large, the rendered curve is hard to interpret so
# I am going to slice different sections of the depth curve to get
# proper GR interpretation
```

```python
import matplotlib.pyplot as plt

# Filter depth range (optional)
depth_filtered = well.loc[0:13000]  # keep this if you want a depth slice

# Extract the GR log
gr = depth_filtered['GR']

# Create the plot
plt.figure(figsize=(6, 12))  # Taller than wide to make it vertical
plt.plot(gr, depth_filtered.index, color='green')  # x = GR, y = Depth
plt.gca().invert_yaxis()  # Depth increases downward
plt.xlabel('Gamma Ray (API)')
plt.ylabel('Depth')
plt.title('Greenland Basin Gamma Ray (API)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Slice from depth 0 to 2000
well.loc[0:500, 'GR'].plot(
    figsize=(10, 4),
    title='Greenland Basin Gamma Ray (API)',
    ylabel='GR (API)'
)
```

```python
well['GR'].plot(kind='hist', bins=30, color='blue', alpha=0.5, density=True, edgecolor='black')
well['GR'].plot(kind='kde', color='black')
plt.xlabel('Gamma Ray - API', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.xlim(0,200)
```

```python
mean = well['GR'].mean()
p5 = well['GR'].quantile(0.05)
p95 = well['GR'].quantile(0.95)

print(f'Mean: \t {mean}')
print(f'P05: \t {p5}')
print(f'P95: \t {p95}')
```

```python
well['GR'].plot(kind='hist', bins=30, color='blue', alpha=0.5, edgecolor='black')
plt.xlabel('Gamma Ray', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xlim(0,200)

plt.axvline(mean, color='blue', label='mean')
plt.axvline(p5, color='green', label='5th Percentile')
plt.axvline(p95, color='purple', label='95th Percentile')

plt.legend()
plt.show()
```
